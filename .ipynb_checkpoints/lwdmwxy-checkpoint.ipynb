```python
###########  网络
def mynetwork(mydata,mynodes,t,b):    ##数据集 ，节点集，时间间隔，滑动多少
    dt = t * 20
    AA = np.empty((ij_n,ij_n))
    G = nx.Graph()
    G.add_nodes_from(mynodes['lable'])
    for i in range(len(mydata)):
        if mydata['time'][i] >= mydata['time'][0]+b:
            if mydata['time'][i] <= mydata['time'][0]+b+(t * 20):
                G.add_edge(mydata['i'][i],mydata['j'][i])
            else:
                break
#        AA = nx.to_numpy_matrix(G) #邻接矩阵
    AA = np.array(nx.adjacency_matrix(G).todense())
    return AA
```

```python
def mypayoff(A,P,S,N):  ###邻接矩阵 博弈矩阵 策略矩阵 收益矩阵 整个网络第N次博弈
    RR = np.zeros(len(A))
    for i in range(len(A)):
        for j in range(len(A)):
            if A[i,j] != 0:
                if S[N,i] == 0 :
                    if S[N,j] == 0:
                        RR[i] = RR[i] + P[0,0] 
                    else:
                        RR[i] = RR[i] + P[0,1]
                else :
                    if S[N,j] == 0:
                        RR[i] = RR[i] + P[1,0] 
                    else:
                        RR[i] = RR[i] + P[1,1]
    return RR
```

```python
#### 博弈次数
def gamen(A,gn):   ###邻接矩阵，博弈次数矩阵
    for i in range(len(A)):
        if np.count_nonzero(A[i]) != 0:
            gn[i] = gn[i] + 1
    return gn
#gn =  gamen(A,np.zeros(len(A)))
#print(gn)
#np.count_nonzero(A[i])
#str(A[i]).count('1')
```

```python
###  策略更新   历史 + 模仿
def Supdata_l(gn,R_c,R_h,w,lmd):    ####策略集，收益矩阵，个体博弈次数，个体标签
    coor_p = 0
    nmd = 1-(2/(math.exp(gn*lmd)+math.exp(-gn*lmd)))
    if R_h == 0:
        coor_p = (1-nmd)*w
    else:
        coor_p = nmd*(R_c/R_h)+(1-nmd)*w
    return coor_p
    
```

```python
###  策略更新   情绪 + 模仿
def Supdata_5(gn,nlable,w,p1,p2):    ####策略集，收益矩阵，个体博弈次数，个体标签
    coor_p = 0
    coor_p = w*p2+p1*nlable
    return coor_p
    
```

```python
###  策略更新   历史 + 情绪
def Supdata_4(gn,nlable,R_c,R_h,p1,p2):    ####策略集，收益矩阵，个体博弈次数，个体标签
    coor_p = 0
    if R_c == 0 or R_h == 0:
        coor_p = p1*nlable
    else:
        coor_p = (R_c/R_h)*p2+p1*nlable
    return coor_p
```

```python
def Supdata_3(gn,R_c,R_h,w):   ###仅历史 
    coor_p = 0
    if R_h == 0:
        coor_p = 0
    else:
        coor_p = R_c/R_h
    return coor_p
    
```

```python
###  策略更新   综合
def Supdata_2(gn,nlable,R_c,R_h,w,p1,p2,lmd):    ####策略集，收益矩阵，个体博弈次数，个体标签
    coor_p = 0
    nmd = 1-(2/(math.exp(gn*lmd)+math.exp(-gn*lmd)))
    if R_c == 0 or R_h == 0:
        coor_p = p2*(1-nmd)*w+p1*nlable  
    else:
        coor_p = (nmd*(R_c/R_h) +(1-nmd)*w)*p2+p1*nlable
    return coor_p
    
```

```python
def wij(R,mydata,mynodes,t,b):     ###i随机模仿一个邻居  l表示游戏轮数
    i_linju = np.zeros(len(mynodes))
    linju_n = np.zeros(len(mynodes))
    mf_j = np.zeros(len(mynodes))
    dt = t * 20
    G = nx.Graph()
    G.add_nodes_from(mynodes['lable'])
    for i in range(len(mydata)):
        if mydata['time'][i] >= mydata['time'][0]+b:
            if mydata['time'][i] <= mydata['time'][0]+b+(t * 20):
                G.add_edge(mydata['i'][i],mydata['j'][i])
            else:
                break
    A = np.array(nx.adjacency_matrix(G).todense())
    for l in range(len(A)):
        if np.count_nonzero(A[l]) == 0:
            mf_j[l] = -1
        else:
            SN = G[l]
            i_linju = list(SN.keys())
#        linju_n[l] = np.sum(A[l]==1)  #l邻居个数
#        for k in SN.keys():
#            for j in range(linju_n[i]):
#                i_linju[j] = k
            mf_j[l] = random.choice(i_linju)  #随机选择i的一个邻居
    return mf_j
#### mf_j  所有节点随机选取邻居的结果
```

```python
def wij1(R,mydata,mynodes,t,b,N):     ###i模仿收益最大的邻居  l表示游戏轮数
    i_linju = np.zeros(len(mynodes))
    linju_n = np.zeros(len(mynodes))
    max_R_linju = np.zeros(len(mynodes))
    R_linju = np.zeros(len(mynodes))
    mf_j = np.zeros(len(mynodes))
    dt = t * 20
    G = nx.Graph()
    G.add_nodes_from(mynodes['lable'])
    for i in range(len(mydata)):
        if mydata['time'][i] >= mydata['time'][0]+b:
            if mydata['time'][i] <= mydata['time'][0]+b+(t * 20):
                G.add_edge(mydata['i'][i],mydata['j'][i])
            else:
                break
    A = np.array(nx.adjacency_matrix(G).todense())
    for l in range(len(A)):
        if np.count_nonzero(A[l]) == 0:
            max_R_linju[l] = -1
        
        else:
            SN = G[l]
            i_linju = list(SN.keys())
            for j in range(len(i_linju)):
                R_linju[j] =  R[N][j]
            max_R_linju[l] = R_linju[1]
            for k in range(len(i_linju)):
                if R_linju[k] >=  max_R_linju[l]:
                    max_R_linju[l] = R_linju[k]
                    mf_j[l] =  i_linju[k]
                else:
                    mf_j[l] =  mf_j[l] 
    return mf_j
#### mf_j  所有节点随机选取邻居的结果
```

```python
def wij2(R,mydata,mynodes,t,b,N):     ###i模仿离差小的邻居  l表示游戏轮数
    l_linju = np.zeros(len(mynodes))
    linju_n = np.zeros(len(mynodes))
    min_R_linju = np.zeros(len(mynodes))
    l_linju_R = np.zeros(len(mynodes))
    lc_linju = np.zeros(len(mynodes))
    mf_j = np.zeros(len(mynodes))
    sum_l_linju_R = 0
    dt = t * 20
    G = nx.Graph()
    G.add_nodes_from(mynodes['lable'])
    for i in range(len(mydata)):
        if mydata['time'][i] >= mydata['time'][0]+b:
            if mydata['time'][i] <= mydata['time'][0]+b+(t * 20):
                G.add_edge(mydata['i'][i],mydata['j'][i])
            else:
                break
    A = np.array(nx.adjacency_matrix(G).todense())
    for l in range(len(A)):
        if np.count_nonzero(A[l]) == 0:
            mf_j[l] = -1
        else:
            SN = G[l]
            l_linju = list(SN.keys())        #l的邻居
            for j in range(len(l_linju)):
                l_linju_R[j] =  R[N][j]     #l的邻居的收益
                sum_l_linju_R = sum_l_linju_R + l_linju_R[j]
            R_mm_linju = sum_l_linju_R/len(l_linju)   #l的邻居收益的均值
            for v in range(len(l_linju)):
                lc_linju[v] = (l_linju_R[j]-R_mm_linju)**2   #l的邻居收益的离差
            min_R_linju[l] = lc_linju[1]
            for k in range(len(l_linju)):  
                if lc_linju[k] <=  min_R_linju[l]:
                    min_R_linju[l] = lc_linju[k]
                    mf_j[l] =  l_linju[k]
                else:
                    mf_j[l] =  mf_j[l] 
    return mf_j
#### mf_j  所有节点选取邻居的结果
```

```python
def wr(R,w,N,S):
    uxy = np.zeros(ij_n)
    w_r = np.zeros(ij_n)
    for i in range(ij_n):
        if w[i] == -1:
            w_r[i] == 0
        else:
            uxy[i] = math.exp((R[N][i]-R[N][int(w[i])])*10)
            if S[N][w[i].astype(int)] == 0 and S[N][i] == 0:
                w_r[i] = 1
            elif S[N][w[i].astype(int)] == 1 and S[N][i] == 1:
                w_r[i] = 0
            elif S[N][w[i].astype(int)] == 0 and S[N][i] == 1:   
                w_r[i] = 1/(1+uxy[i])
            elif S[N][w[i].astype(int)] == 1 and S[N][i] == 0:
                w_r[i] = uxy[i]/(1+uxy[i])
    return w_r
```

```python
#####将合作概率转为策略矩阵
def stra(coor,N,S,A):
    for i in range(len(A)):
        S[N][i] = np.random.binomial(n=1, p=(1-coor[i]))
    return S
def stra1(coor,N,S,n):
    for i in range(n):
        S[N][i] = np.random.binomial(n=1, p=(1-coor[i]))
    return S
```

```python
def upfunction(S,N,A):
    up = 0
    N_lj = np.zeros(len(A))
    N_c = np.zeros(len(A))
    up1 = 0
    for k in range(len(A)):
        if np.count_nonzero(A[k]) != 0:
            N_lj[N] = N_lj[N]+1  ##参与博弈的人数
            if S[N,k] == 0:
                N_c[N] = N_c[N]+1  ##博弈中选择合作的人数
    up = N_c[N]/N_lj[N]  ##### 合作者的比例      
    return up
```

```python
def upfunction1(S,N,A):
    up = np.zeros(len(A))
    N_lj = np.zeros(len(A))
    N_c = np.zeros(len(A))
    for k in range(len(A)):
        N_lj[k] = np.count_nonzero(A[k,])  ##与k博弈的人数
        if N_lj[k] != 0:
            for h in range(len(A)):
                if A[k,h] != 0:
                    if S[N,h] == 0:
                        N_c[k] = N_c[k]+1
                    else:
                        N_c[k] = N_c[k]
            up[k] = N_c[k]/N_lj[k]
        else:
            up[k] = 0 
    return up
```

```python
###4综合  个人    情绪+模仿
y = pd.read_csv('E:/Desktop/wxy/结果/4情绪/x.csv',header = None)
data_n = 50
ij_n = len(ij)
cp = np.zeros((data_n+1,ij_n))
S = np.zeros((data_n+1,ij_n))
cp0 = np.transpose(y[0])
cp0.values.tolist()
cp[0,] = cp0
S = stra1(cp[0,],0,S,ij_n)


for mm in [2]:
    b = 0
    R = np.zeros((data_n,ij_n))                       
    RR = np.zeros((data_n,ij_n))
    R_c = np.zeros(ij_n)
    R_h = np.zeros(ij_n)
    gn = np.zeros(ij_n)
    N_lj = np.zeros(data_n)
    N_c = np.zeros(data_n)
    P_c = np.zeros(data_n)
    bbb = 1 + mm /10
    P = np.array([[1,0],[bbb,0]])
    t = 90   ###一次选取10分钟的数据
    da = 90 #滑动3个窗口，即1分钟
    for N in range(data_n):
        A = mynetwork(SFHH,ij,t,b)
        RR = mypayoff(A,P,S,N)
        R[N,] = RR.reshape(1,ij_n)
        gn = gamen(A,gn)
        for i in range(len(A)):
            if S[N,i] == 0:
                R_c[i] = R[N,i] + R_c[i]
        R_h = np.sum(R,keepdims = True,axis = 0)
        for k in range(len(A)):
            if np.count_nonzero(A[k]) != 0:
                N_lj[N] = N_lj[N]+1  ##参与博弈的人数
                if S[N,k] == 0:
                    N_c[N] = N_c[N]+1  ##博弈中选择合作的人数
        if N_lj[N] == 0:
            P_c[N] = -1    #####没有连接
        else:
            P_c[N] = N_c[N]/N_lj[N]  ##### 合作者的比例

        w_lj = wij2(R,SFHH,ij,t,b,N)    ####所有节点选择的邻居集
        w = wr(R,w_lj,N,S)
        up = upfunction1(S,N,A) 
        uup = 1-up
        for j in range(len(A)):
            if np.count_nonzero(A[j]) == 0:
                cp[N+1,j] = cp[N,j]
            else:               
                cp[N+1,j] = w[j]*uup+up*cp[0,j]
        S = stra(cp[N+1,],N+1,S,A)
        b = 20 *da *N
    allDir = 'b = '+str(mm)
    filenames_out = 'E:/Desktop/wxy/结果/8情绪+模仿/'
    domain2 = os.path.abspath(filenames_out)  # 处理完文件保存地址
    fname2 = ('P_c.csv','hb.csv','S.csv','cp.csv','R.csv')
    RRR = pd.DataFrame(P_c)
    RRR.to_csv(os.path.join(domain2, allDir,fname2[0]))
    zh = [gn,N_lj,N_c]
    ggn = pd.DataFrame(zh)
    ggn.to_csv(os.path.join(domain2, allDir, fname2[1]))
    ss = pd.DataFrame(S)
    ss.to_csv(os.path.join(domain2, allDir, fname2[2]))
    ccp = pd.DataFrame(cp)
    ccp.to_csv(os.path.join(domain2, allDir, fname2[3]))
    RR = pd.DataFrame(R)
    RR.to_csv(os.path.join(domain2, allDir, fname2[4]))
```
