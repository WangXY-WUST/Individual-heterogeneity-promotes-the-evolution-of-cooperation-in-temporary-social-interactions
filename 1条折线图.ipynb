```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
import random
import os
import openpyxl
```

```python
data = pd.read_csv('E:/Desktop/wxy/结果/data.csv',header = None)
filenames_out = 'E:/Desktop/wxy/结果/3lamda/30min/ls'
i = 1 
for mm in range(0,11,2):
    allDir = 'b = ' + str(mm)
    fname2 = 'P_c.csv'
    dataa = pd.read_csv(os.path.join(filenames_out, allDir,fname2))
    data[i] = dataa['0']
    i = i + 3
```
