```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
data = pd.read_csv('E:/Desktop/wxy/结果/data.csv',header = None)
```

```python
filenames_out = ('E:/Desktop/wxy/结果/2模仿/3随机',
                 'E:/Desktop/wxy/结果/2模仿/1收益最大',
                 'E:/Desktop/wxy/结果/2模仿/2方差最小',                           
                )
ts = len(filenames_out)
i = 1
for m in [2,8]:
    for j in range(ts):
        allDir = 'b = ' + str(m)
        fname2 = 'P_c.csv'
        dataa = pd.read_csv(os.path.join(filenames_out[j], allDir,fname2))
        data[i] = dataa['0']
        i = i + 1 
```
