```python
#from sympy import *
#linecolor = ('#d67274','#b1c769','#ff2468','#0f7601','#C83828','#005b63')

linecolor = ('#d52685','#6cbc35','#553a99','#fd9f3e','#075aaa')
plt.style.use('science')
xx=Symbol("xx")


n = 20
x = np.zeros(n,dtype=np.float32)
y1 = np.zeros(n,dtype=np.float32) 
y2 = np.zeros(n,dtype=np.float32)
y3 = np.zeros(n,dtype=np.float32)
y4 = np.zeros(n,dtype=np.float32)
y5 = np.zeros(n,dtype=np.float32)

z1 = np.zeros(n,dtype=np.float32) 
z2 = np.zeros(n,dtype=np.float32)
z3 = np.zeros(n,dtype=np.float32)
z4 = np.zeros(n,dtype=np.float32)
z5 = np.zeros(n,dtype=np.float32)

for i in range(n):
    x[i] = i
    y1[i] = 2/(math.exp(i)+math.exp(-i))
    y2[i] = 2/(math.exp(0.8*i)+math.exp(-i*0.8))
    y3[i] = 2/(math.exp(0.6*i)+math.exp(-i*0.6))
    y4[i] = 2/(math.exp(0.4*i)+math.exp(-i*0.4))
    y5[i] = 2/(math.exp(0.2*i)+math.exp(-i*0.2))
    
    
for i in range(n):
    x[i] = i
    z1[i] = -(2*(math.exp(i)-math.exp(-i)))/((math.exp(i)+math.exp(-i))**2)
    z2[i] = -(2*(math.exp(0.8*i)-math.exp(-i*0.8)))/((math.exp(0.8*i)+math.exp(-i*0.8))**2)
    z3[i] = -(2*(math.exp(0.6*i)-math.exp(-i*0.6)))/((math.exp(0.6*i)+math.exp(-i*0.6))**2)
    z4[i] = -(2*(math.exp(0.4*i)-math.exp(-i*0.4)))/((math.exp(0.4*i)+math.exp(-i*0.4))**2)
    z5[i] = -(2*(math.exp(0.2*i)-math.exp(-i*0.2)))/((math.exp(0.2*i)+math.exp(-i*0.2))**2)

plt.figure(figsize=(6,2),dpi=600)
plt.subplots_adjust(wspace=0.4)
font_size = 10

plt.subplot(1,2,1) 
plt.plot(x,y5, color=linecolor[4],label ='m = 0.2',linewidth = 1,linestyle = '-',marker = 'D',markersize = 3,markeredgecolor = linecolor[4])
plt.plot(x,y4, color=linecolor[3],label ='m = 0.4',linewidth = 1,linestyle = '-',marker = 'p',markersize = 3,markeredgecolor = linecolor[3])
plt.plot(x,y3, color=linecolor[2],label ='m = 0.6',linewidth = 1,linestyle = '-',marker = 'v',markersize = 3,markeredgecolor = linecolor[2])
plt.plot(x,y2, color=linecolor[1],label ='m = 0.8',linewidth = 1,linestyle = '-',marker = 's',markersize = 3,markeredgecolor = linecolor[1])
plt.plot(x,y1, color=linecolor[0],label ='m = 1.0',linewidth = 1,linestyle = '-',marker = 'o',markersize = 3,markeredgecolor = linecolor[0])


plt.title("(a) "+ r'$\lambda$'+ '(mn)')
plt.legend(ncol=1, loc='upper right',frameon = False,labelspacing=0.2 )
plt.xlabel('n',)
plt.ylabel(r'$\lambda$'+ '(mn)')
plt.xticks((0,5,10,15,20))

plt.subplot(1,2,2)
plt.plot(x,z1, color=linecolor[0],label ='m = 1.0',linewidth = 1,linestyle = '-',marker = 'o',markersize = 3,markeredgecolor = linecolor[0] )
plt.plot(x,z2, color=linecolor[1],label ='m = 0.8',linewidth = 1,linestyle = '-',marker = 's',markersize = 3,markeredgecolor = linecolor[1])
plt.plot(x,z3, color=linecolor[2],label ='m = 0.6',linewidth = 1,linestyle = '-',marker = 'v',markersize = 3,markeredgecolor = linecolor[2])
plt.plot(x,z4, color=linecolor[3],label ='m = 0.4',linewidth = 1,linestyle = '-',marker = 'p',markersize = 3,markeredgecolor = linecolor[3])
plt.plot(x,z5, color=linecolor[4],label ='m = 0.2',linewidth = 1,linestyle = '-',marker = 'D',markersize = 3,markeredgecolor = linecolor[4])
plt.xlabel('n')
plt.ylabel("d" + r'$\lambda$'+ "(mn)/dn")
#plt.plot(x,y2, color='#1B5E20',label = r'$\lambda$'+ '(n/2)',linewidth = 1)
#plt.plot(x,y3, color='#4d5aaf',label = 'd'+r'$\lambda$'+ "/dn",linewidth = 1,linestyle = '--')
plt.title("(b) d"+r'$\lambda$'+ "(mn)/dn")
#plt.yticks((0,1))#设置大小及加粗
plt.xticks((0,5,10,15,20))
plt.savefig('E:/Desktop/毕业论文/论文1/202307结果/figures/figure1.svg',bbox_inches='tight',pad_inches=0.0)
#plt.savefig('E:/Desktop/epl/epl论文格式模板/epl2/figure1.eps',bbox_inches='tight',pad_inches=0.0)
#plt.savefig('E:/Desktop/wxy/论文图/figure1.svg',bbox_inches='tight',pad_inches=0.0)
```
