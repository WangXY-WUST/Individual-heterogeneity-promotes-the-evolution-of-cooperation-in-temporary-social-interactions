```python
#-*- encoding=utf-8 -*- 
#import matplotlib zhfont1 = matplotlib.font_manager.FontProperties(fname='C:\Windows\Fonts\simsun.ttc') 
# 按照固定区间长度绘制频率分布直方图 
# bins_interval 区间的长度 
# margin 设定的左边和右边空留的大小 
def probability_distribution(data, bins_interval=1, margin=1): 
    plt.figure(dpi=600,figsize=(3,2))
    font_size = 6
    axl = plt.subplot(1,1,1) 
    bins = range(min(data), max(data) + bins_interval - 1, bins_interval) 
    print(len(bins)) 
    for i in range(0, len(bins)): 
        print(bins[i]) 
    plt.xlim(0,50) 
    plt.ylim(0,30)
    plt.xlabel('N',fontdict = {'family' : 'Times New Roman','size': font_size,'style':'oblique'}) 
    plt.ylabel('Number of nodes',font = 'Times New Roman',size = font_size) 
    q1n = int(data.min())
    q1x = int(data.max())
#    plt.xticks([q1n,q1n+int((q1x-q1n)/5),q1n+int(((q1x-q1n)/5)*2),q1n+int(((q1x-q1n)/5)*3),
#               q1n+int(((q1x-q1n)/5)*4),q1x])
    plt.xticks([0,5,10,15,20,25,30,35,40,45,50])
    plt.yticks([0,5,10,15,20,25,30,])
    plt.rc('font',family = 'Times New Roman' )
    # 频率分布normed=True，频次分布normed=False 
    prob,left,rectangle = plt.hist(x=data, bins=bins, histtype='bar',rwidth=0.9) 
    for x, y in zip(left, prob): 
        # 字体上边文字 
        # 频率分布数据 
        normed=True 
        plt.text(x + bins_interval / 2, y + 0.003, '%.0f' % y,
                 ha='center', va='bottom',font = 'Times New Roman',size = font_size)
        # 频次分布数据 normed=False 
        # plt.text(x + bins_interval / 2, y + 0.25, '%.2f' % y, ha='center', va='top')
    labels = axl.get_xticklabels() + axl.get_yticklabels()
    [label.set_fontname('Times New Roman') for label in labels]
    plt.tick_params(labelsize=font_size)
    plt.savefig('E:/Desktop/epl/epl论文格式模板/epl2/figure4.eps',bbox_inches='tight',pad_inches=0.0)
    plt.savefig('E:/Desktop/wxy/论文图/figure5.svg',bbox_inches='tight',pad_inches=0.0)
    plt.show()
```

```python
data = gn[0]

probability_distribution(data = data,bins_interval = 1,margin = 0.9)
```
