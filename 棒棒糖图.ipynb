```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('E:/Desktop/wxy/结果/网络特征.csv',header = None)
```

```python
# custom the stem lines
plt.figure(dpi=600,figsize=(3.14,1))
font_size = 4
plt.subplots_adjust(hspace=0, wspace=0.4)
ax1 = plt.subplot(1,2,1)
(markerline, stemlines, baseline) = plt.stem(data[0], data[1],linefmt= '-',
                                             bottom = 0, use_line_collection = True, 
                                             basefmt='y--')
plt.setp(stemlines, linestyle="-", color="k", linewidth=0.5 )
plt.setp(markerline, marker='o', markersize=1.2, markeredgecolor="#FF6347")
markerline.set_markerfacecolor('#FF6347') 
plt.xlabel('Number of nodes',font = 'Times New Roman',size = font_size)
plt.ylabel('Number of networks',font = 'Times New Roman',size = font_size)
plt.title('(a)',font = 'Times New Roman',size = font_size,y=-0.6)
plt.xticks([min(data[0]),100,200,300,max(data[0])])
plt.yticks([0,1,2,3])
labels = ax1.get_xticklabels() + ax1.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]
plt.tick_params(labelsize=font_size)

ax2 = plt.subplot(1,2,2)
(markerline, stemlines, baseline) = plt.stem(data[2], data[3],linefmt= '-',
                                             bottom = 0, use_line_collection = True, 
                                             basefmt='y--')
plt.setp(stemlines, linestyle="-", color="k", linewidth=0.5 )
plt.setp(markerline, marker='o', markersize=1.2, markeredgecolor="#FF6347")
markerline.set_markerfacecolor('#FF6347') 
plt.xlabel('Number of edges',font = 'Times New Roman',size = font_size)
plt.ylabel('Number of networks',font = 'Times New Roman',size = font_size)
plt.xticks([min(data[2]),500,1000,1500,max(data[2])])
plt.yticks([0,1,2,3])
plt.title('(b)',font = 'Times New Roman',size = font_size,y=-0.6)
labels = ax2.get_xticklabels() + ax2.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]
plt.tick_params(labelsize=font_size)


plt.savefig('E:/Desktop/wxy/结果/figure4.eps')
plt.savefig('E:/Desktop/wxy/论文图/figure4.svg',bbox_inches='tight',pad_inches=0.0)
```
